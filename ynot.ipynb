```python
import json

def wriylop():
    ggh = []
    for i in range(len(result.boxes)):
        boxes = result.boxes[i]
        classnk = names[result.boxes.cls[i].item()]
        boxes.xyxy[0][1]
        aspa = {
            "class_name": classnk,
            "bounding_box": {
                "xmin": boxes.xyxy[0][0].item(),
                "ymin": boxes.xyxy[0][1].item(),
                "xmax": boxes.xyxy[0][2].item(),
                "ymax": boxes.xyxy[0][3].item()
            }
        }
        
        ggh.append(aspa) 
        print(ggh)

    data = {
        "image_path": "path/to/image.jpg",
        "image_width": width,
        "image_height": height,
        "objects": ggh
    }
    # Convert the JSON data to a string
    json_str = json.dumps(data, indent=4)

    return json_str  # Return the JSON string instead of writing it to a file

# Run batched inference on a list of images
results = model("uploads/",save=True, stream=True, conf=0.4) # return a list of Results objects
print(results)
names = model.names
print(names)

# Initialize json_str as an empty string
json_str = ""

for result in results:
    print("hello")
    # boxes = result.boxes  # Boxes object for bbox outputs
    print("boxes", result.boxes.xyxy, "shape", result.boxes.xyxy.size())
    print("confidence", result.boxes.conf)  # confidence score, (N, 1)
    print("results", result.boxes.cls)    # cls, (N, 1)
    
    # Concatenate each JSON string to the existing json_str
    json_str += wriylop()

# Print the final JSON string
print(json_str)
```
